**Autograd** in PyTorch is a core feature that **automatically computes gradients (derivatives)** for tensors used in neural networks.

## üîç Short Explanation:
- It enables **automatic backpropagation** by **tracking operations** on tensors.
- When you call **`.backward()`** on a loss tensor, PyTorch **calculates all gradients** needed for optimization.
- This eliminates the need to **manually compute derivatives**.

‚úÖ **Why it's useful**:
- Saves time  
- Reduces errors  
- Essential for training deep learning models
